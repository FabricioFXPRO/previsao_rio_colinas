```python
import pandas as pd
import folium

# Carregar o dataframe das estações

df = pd.read_csv('Dados_Bacia_8_Subestacao_86_Ativas.csv', sep=';')

# Lista dos códigos das estações que você quer selecionar
codigos_estacoes = ['86160000', '86306000', '86510000', '86743800', '86195000', 
                    '86390900', '86490500', '86406000', '86117000', '86100600']

# Converter a coluna 'codigoestacao' para string, garantindo que a comparação seja correta
df['codigoestacao'] = df['codigoestacao'].astype(str)

# Filtrar o dataframe para as estações com os códigos selecionados
df_selecionadas = df[df['codigoestacao'].isin(codigos_estacoes)]

# Função para formatar o popup com informações adicionais
def format_popup(row):
    return (f"Estação: {row['UF_Nome_Estacao']}<br>"
            f"Código da Estação: {row['codigoestacao']}<br>"
            f"Latitude: {row['Latitude']}<br>"
            f"Longitude: {row['Longitude']}")

# Criar o mapa base centrado em uma localização aproximada
mapa = folium.Map(location=[-29.5, -51.9], zoom_start=10)

# Adicionar as estações selecionadas ao mapa
for _, row in df_selecionadas.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=format_popup(row),
        icon=folium.Icon(color='blue')
    ).add_to(mapa)

# Salvar o mapa em um arquivo HTML
mapa.save('mapa_estacoes_selecionadas.html')

print("Mapa salvo como 'mapa_estacoes_selecionadas.html'")
```
