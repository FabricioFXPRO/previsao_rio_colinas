```python
import requests

# Função para obter o token de acesso
def get_token():
    url = "https://www.ana.gov.br/hidrowebservice/EstacoesTelemetricas/OAUth/v1"
    
    headers = {
        "accept": "*/*",
        "Identificador": "LOGIN",  # Substitua com o seu identificador
        "Senha": "SENHA"  # Substitua com a sua senha
        # O loguin e senha foram subistiruido para segurança
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        # Retorna o token de autenticação
        data = response.json()
        token = data["items"]["tokenautenticacao"]
        return token
    else:
        raise Exception(f"Erro ao obter token: {response.status_code}, {response.text}")

# Função para fazer uma requisição com um token, renovando-o se necessário
def make_request(url, token):
    headers = {
        "Authorization": f"Bearer {token}",
        "accept": "*/*"
    }

    # Fazer a requisição
    response = requests.get(url, headers=headers)

    # Se o token tiver expirado (erro 401), gera um novo token e tenta novamente
    if response.status_code == 401:
        print("Token expirado, gerando um novo token...")
        token = get_token()
        headers["Authorization"] = f"Bearer {token}"
        response = requests.get(url, headers=headers)

    # Se ainda houver erro, mostrar a mensagem de erro
    if response.status_code != 200:
        print(f"Erro na requisição: {response.status_code}, mensagem: {response.text}")
    else:
        # Retorna os dados se a requisição for bem-sucedida
        return response.json()

# Exemplo de uso:
if __name__ == "__main__":
    # Obter o primeiro token
    token = get_token()

    # URL do endpoint desejado (substitua pelo endpoint que você quer acessar)
    url = "https://www.ana.gov.br/hidrowebservice/EstacoesTelemetricas/HidroSubBacia/v1"

    # Fazer a requisição
    data = make_request(url, token)

    # Exibir os dados retornados (se houver)
    if data:
        print(data)
```
