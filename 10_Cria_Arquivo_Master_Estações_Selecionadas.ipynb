```python
import pandas as pd

# Listas de arquivos e códigos
arquivos_cota_vazao = ['86510000_Cota-Vazão_FINAL.csv']
codigos_estacoes_cota = ['86510000']

arquivos_chuva = [
    '86743800_chuva_adotada_FINAL.csv', '86510000_chuva_adotada_FINAL.csv',
    '86490500_chuva_adotada_FINAL.csv', '86406000_chuva_adotada_FINAL.csv',
    '86390900_chuva_adotada_FINAL.csv', '86306000_chuva_adotada_FINAL.csv',
    '86195000_chuva_adotada_FINAL.csv', '86160000_chuva_adotada_FINAL.csv',
    '86117000_chuva_adotada_FINAL.csv', '86100600_chuva_adotada_FINAL.csv'
]
codigos_estacoes_chuva = ['86743800', '86510000', '86490500', '86406000', '86390900',
                          '86306000', '86195000', '86160000', '86117000', '86100600']

# Dataframe mestre inicial
df_master = None

# Função para processar e renomear dataframes
def processar_dataframe(arquivo, codigo, colunas_interesse, prefixo):
    df = pd.read_csv(arquivo)
    df = df[['Data_Hora_Medicao'] + colunas_interesse]
    novos_nomes = {coluna: f"{coluna}_{codigo}" for coluna in colunas_interesse}
    df.rename(columns=novos_nomes, inplace=True)
    return df

# Processar arquivos de cota e vazão
for arquivo, codigo in zip(arquivos_cota_vazao, codigos_estacoes_cota):
    df = processar_dataframe(
        arquivo, codigo, ['Cota_Adotada', 'Vazao_Adotada'], 'Cota_Vazao'
    )
    if df_master is None:
        df_master = df
    else:
        df_master = pd.merge(df_master, df, on='Data_Hora_Medicao', how='outer')

# Processar arquivos de chuva
for arquivo, codigo in zip(arquivos_chuva, codigos_estacoes_chuva):
    df = processar_dataframe(arquivo, codigo, ['Chuva_Adotada'], 'Chuva')
    df_master = pd.merge(df_master, df, on='Data_Hora_Medicao', how='outer')

# Converter 'Data_Hora_Medicao' para datetime
df_master['Data_Hora_Medicao'] = pd.to_datetime(df_master['Data_Hora_Medicao'])

# Ordenar por data e hora
df_master.sort_values('Data_Hora_Medicao', inplace=True)

# Resetar o índice
df_master.reset_index(drop=True, inplace=True)
```

```python

#faz um relatorio do numero de valores nulos de cada coluna
df_master.isnull().sum()
```

```python
#Preenche os valores nulos com 0
df_master.fillna(0, inplace=True)
```

```python

#faz um relatorio do numero de valores nulos de cada coluna
df_master.isnull().sum()
```

```python
# Salva o dataframe mestre
df_master.to_csv('0001_Master.csv', index=False)
df_master
```
